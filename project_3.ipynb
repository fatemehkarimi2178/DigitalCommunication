BPSK Modulation in Rayleigh Fading Channel


```matlab
clear all
close all
clc
iteration=10^5;
snrdb=[-4:10];
simsnr=10.^(snrdb/10);
erpr=zeros(1,length(snrdb))
for jjj=1:length(snrdb)
    snr=10^(snrdb(jjj)/10);
    count=0;
    for i=1:iteration
        bit=round(rand(1));
        data=2*bit-1;
        H = sqrt(.5)*( randn(1) + j.*randn(1));
        noise=randn(1);
        received=(sqrt(.5)*noise) + sqrt(snr)*abs(H)*data;
        if received >=0
            detected=1;
        else 
        detected=0;
        end
        if ((detected==0 && bit==1)||(detected==1 && bit==0))
        count=count+1;
        end
    end
    erpr(jjj)=count/iteration;
end
erpr
qf=qfunc(sqrt(2.*simsnr));
semilogy(snrdb,erpr,'-r*',snrdb,qf,'-o'); 
xlabel('E_b/N_0')
ylabel('Bit Error Probability')
grid on
```
